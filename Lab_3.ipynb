```python
# -- coding: utf-8 --
# Script: Calculo paso a paso de Jitter y Shimmer (sin definir funciones)
# Entrada: listas de periodos T (segundos) y amplitudes A (unidades relativas) por grabación
# Salida: imprime desarrollo paso a paso y valores numéricos

y 'A' (amplitudes)
grabaciones = [
    { 'name': 'Hombre 1', 'T': [0.0098, 0.0097, 0.0099, 0.00985, 0.0098], 'A': [0.8, 0.82, 0.79, 0.81, 0.80] },
    { 'name': 'Hombre 2', 'T': [0.0101, 0.0100, 0.00995, 0.01005], 'A': [0.75, 0.76, 0.74, 0.755] },
    { 'name': 'Hombre 3', 'T': [0.0110, 0.0109, 0.01105, 0.01095, 0.011], 'A': [0.9, 0.88, 0.91, 0.89, 0.90] },
    { 'name': 'Mujer 1',  'T': [0.0068, 0.00685, 0.00682, 0.00680], 'A': [0.6, 0.62, 0.61, 0.605] },
    { 'name': 'Mujer 2',  'T': [0.0065, 0.00655, 0.00652, 0.00650, 0.00651], 'A': [0.7, 0.695, 0.702, 0.699, 0.701] },
    { 'name': 'Mujer 3',  'T': [0.0072, 0.00718, 0.00722, 0.00719], 'A': [0.85, 0.86, 0.84, 0.855] }
]


print("\n=== MEDICIÓN DE JITTER Y SHIMMER (desarrollo paso a paso) ===\n")

for rec in grabaciones:
    nombre = rec['name']
    T = rec['T'][:]  # periodos (lista)
    A = rec['A'][:]  # amplitudes (lista)
    N_T = 0
    N_A = 0
    # Contar elementos manualmente (sin usar len) para mostrar "sin funciones"
    for _ in T:
        N_T += 1
    for _ in A:
        N_A += 1

    print(f"Grabación: {nombre}")
    print(f"Periodo T (lista): {T}")
    print(f"Amplitudes A (lista): {A}")
    print(f"N (periodos) = {N_T}, N (amplitudes) = {N_A}")


    # JITTER ABSOLUTO (desarrollo)


    if N_T < 2:
        print("  No hay suficientes periodos para calcular Jitter.")
    else:
        # construimos lista de diferencias absolutas y cadena con el desarrollo
        diffs_T = []
        terms_T = []  # para mostrar |T1-T2| + |T2-T3| + ...
        for i in range(N_T - 1):
            # diferencia absoluta paso a paso
            diff = T[i] - T[i+1]
            if diff < 0:
                diff_abs = -diff
            else:
                diff_abs = diff
            diffs_T.append(diff_abs)
            # indexación humana: T1, T2, ...
            term_str = f"|T{i+1}-T{i+2}|"
            terms_T.append(term_str)

        # construir la cadena del sumatorio desarrollado
        desarroll_t = " + ".join(terms_T)
        # suma manual de diffs
        suma_diffs_T = 0.0
        for val in diffs_T:
            suma_diffs_T += val

        jitter_abs = suma_diffs_T / (N_T - 1)

        # Imprimir desarrollo
        print("  Jitter_abs (fórmula): Jitter_abs = (1/(N-1)) * Σ_{i=1}^{N-1} |T_i - T_{i+1}|")
        print(f"  Desarrollo: Jitter_abs = ( {desarroll_t} ) / ({N_T} - 1)")
        # mostrar valores numéricos por término
        term_values = " + ".join([f"{v:.8f}" for v in diffs_T])
        print(f"  Valores: Jitter_abs = ( {term_values} ) / {N_T - 1}")
        print(f"  Suma de diferencias = {suma_diffs_T:.10f}")
        print(f"  Jitter_abs = {suma_diffs_T:.10f} / {N_T - 1} = {jitter_abs:.10f}")


        # JITTER RELATIVO (%)
        # Jitter_rel = ( (1/(N-1) * sum |T_i-T_{i+1}| ) / ( (1/N) * sum T_i ) ) * 100

        # suma manual de T_i
        suma_T = 0.0
        for val in T:
            suma_T += val
        periodo_medio = suma_T / N_T

        jitter_rel = (jitter_abs / periodo_medio) * 100.0

        # imprimir desarrollo del jitter relativo
        print("  Jitter_rel (fórmula): Jitter_rel = ( (1/(N-1) Σ |T_i-T_{i+1}| ) / ( (1/N) Σ T_i ) ) * 100")
        print(f"  Desarrollo numérico = ( {suma_diffs_T:.10f}/{N_T - 1} ) / ( {suma_T:.10f}/{N_T} ) * 100")
        print(f"  Periodo medio = (T1 + T2 + ... + TN) / N = {suma_T:.10f} / {N_T} = {periodo_medio:.10f}")
        print(f"  Jitter_rel = {jitter_rel:.8f} %")


    # SHIMMER ABSOLUTO (desarrollo)
    # Shimmer_abs = (1/(N-1)) * sum_{i=1}^{N-1} |A_i - A_{i+1}|

    if N_A < 2:
        print("  No hay suficientes picos de amplitud para calcular Shimmer.")
    else:
        diffs_A = []
        terms_A = []
        for i in range(N_A - 1):
            diff = A[i] - A[i+1]
            if diff < 0:
                diff_abs = -diff
            else:
                diff_abs = diff
            diffs_A.append(diff_abs)
            terms_A.append(f"|A{i+1}-A{i+2}|")

        desarroll_a = " + ".join(terms_A)
        suma_diffs_A = 0.0
        for val in diffs_A:
            suma_diffs_A += val
        shimmer_abs = suma_diffs_A / (N_A - 1)

        print("  Shimmer_abs (fórmula): Shimmer_abs = (1/(N-1)) * Σ_{i=1}^{N-1} |A_i - A_{i+1}|")
        print(f"  Desarrollo: Shimmer_abs = ( {desarroll_a} ) / ({N_A} - 1)")
        term_values_A = " + ".join([f"{v:.8f}" for v in diffs_A])
        print(f"  Valores: Shimmer_abs = ( {term_values_A} ) / {N_A - 1}")
        print(f"  Suma de diferencias amplitud = {suma_diffs_A:.10f}")
        print(f"  Shimmer_abs = {suma_diffs_A:.10f} / {N_A - 1} = {shimmer_abs:.10f}")


        # SHIMMER RELATIVO (%)
             suma_A = 0.0
        for val in A:
            suma_A += val
        amplitude_mean = suma_A / N_A

        shimmer_rel = (shimmer_abs / amplitude_mean) * 100.0

        print("  Shimmer_rel (fórmula): Shimmer_rel = ( (1/(N-1) Σ |A_i-A_{i+1}| ) / ( (1/N) Σ A_i ) ) * 100")
        print(f"  Desarrollo numérico = ( {suma_diffs_A:.10f}/{N_A - 1} ) / ( {suma_A:.10f}/{N_A} ) * 100")
        print(f"  Amplitud media = {suma_A:.10f} / {N_A} = {amplitude_mean:.10f}")
        print(f"  Shimmer_rel = {shimmer_rel:.8f} %")

    print("-" * 70)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
import scipy.signal as sg
import soundfile as sf # Import the soundfile library

WAV_FILE = "/content/drive/MyDrive/Audios lab 3-20251001T191505Z-1-001/Audios lab 3/HOMBRE1.wav"
GENERO   = "hombre"


# Leer archivo
x, fs = sf.read(WAV_FILE)
if x.ndim > 1: x = x[:,0]

# Filtro pasa-banda
bp = (80,400) if GENERO=="hombre" else (150,500)
b,a = sg.butter(4, [bp[0]/(fs/2), bp[1]/(fs/2)], btype="band")
x = sg.filtfilt(b,a,x)

# Envolvente y detección de picos
env = np.abs(sg.hilbert(x))
env = sg.convolve(env, np.ones(int(0.002*fs))/int(0.002*fs), mode="same") # suavizado 2ms
pks,_ = sg.find_peaks(env, distance=int(fs/bp[1]*0.8), prominence=0.25*env.std())

# Períodos (Ti) y amplitudes (Ai)
Ti = np.diff(pks)/fs
Ai = env[pks]

# Jitter y Shimmer (fórmulas de la imagen) (aquí poner las imagines del lab )
jit_abs = np.mean(np.abs(Ti[:-1]-Ti[1:]))
jit_rel = (jit_abs/np.mean(Ti))*100
shim_abs = np.mean(np.abs(Ai[:-1]-Ai[1:]))
shim_rel = (shim_abs/np.mean(Ai))*100

print(f"\n--- Resultados ({WAV_FILE}) ---")
print(f"Jitter absoluto (s): {jit_abs:.6e}")
print(f"Jitter relativo (%): {jit_rel:.4f}")
print(f"Shimmer absoluto:    {shim_abs:.6e}")
print(f"Shimmer relativo (%):{shim_rel:.4f}")

# Gráfico opcional (para entender mejor )
t = np.arange(len(x))/fs
plt.plot(t,x,label="Señal filtrada")
plt.plot(pks/fs,env[pks],"ro",label="Picos")
plt.legend(); plt.xlabel("Tiempo [s]"); plt.ylabel("Amplitud")
plt.title("Detección de ciclos"); plt.grid(); plt.show()
```
